```python
import matplotlib.pyplot as plot
```

```python
def fftfreq(Fs, n):
    f0 = [1] * n

    f1 = [0] * n
    for i in range(int(n/2)):
        f1[i] = i * Fs / n

    i = int(n/2)
    while i < n:
        f1[i] = - ((n-i) * Fs / n)
        i+=1

    f2 = [0]*n
    for i in range(n):
        f2[i] = f1[i] * f1[i]

    return f0, f1, f2
```

```python
Fs = 20000
n = 10
f0, f1, f2 = fftfreq(Fs, n)
```

```python
plot.plot(f0)
f0
```

```python
plot.plot(f1)
f1
```

```python
plot.plot(f2)
f2
```
